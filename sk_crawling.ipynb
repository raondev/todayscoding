```python
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
#명시적 대기를 위해
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
#사전에 정보 로드 DB shell
main_url='http://tour.interpark.com/'
keyword='로마'

#드라이버 로드
#mac  ./chromedriver
driver = wd.Chrome(executable_path='chromedriver.exe')

#사이트 접속(get)
driver.get(main_url)

# 여행 검색창 찾기 및 검색어 입력 (id : SearchGNBText)
#selecter # , id는 그냥  선수로 css selecter 학습 필요
# element 인지 elements인지 확인 해야함 하나인지 두개 이상인지
#수정할경우 -> 뒤에 내용이 붙어버림 -> .clear()  -> send_keys(keyword)

driver.find_element_by_id('SearchGNBText').send_keys(keyword)

#검색버튼 클릭
#search-btn 찾음
#driver.find_element_by_css_selector('.search-btn').click()
driver.find_element_by_css_selector('button.search-btn').click()

#주의 사항 url직접 입력 하여 가면 빠르나 사이트 측이 바꿀수 있음
# 그래서 단계별로 진행 하는 것이 중요함

#잠시대기 -페이지로드후 즉각적으로 데이터 획득은 주의 해야 한다.(사이트 속도 문제로 일낟 기달려야 한다)
#명시적 대기 -> 특정 요소사 로케이트(발견) 대기
try:
    element = WebDriverWait(driver,10).until(
            EC.presence_of_element_located((By.CLASS_NAME,'oTravelBox'))
    )
except Exception as e:
    print('오류발생', e)
    
#암묵적 대기 -> DOM이 다 로드 될때 까지 대기 하고 먼저 로드되면 바로 진행
#요소를 찾을 특정 시간 동안 DOM 풀링을 지시 10초이내라도 발견되면 진행

driver.implicitly_wait(10)

#절대 대기 -> time.sleep(10) - > 크라우드 페어(디도스 방어 솔루션)
#더보기 눌러서 -> 게시판 진입
driver.find_element_by_css_selector('.oTravelBox>.boxList>.moreBtnWrap>.moreBtn').click()

#게시판에서 데이터를 가져올때 데이터가 많으면 세션(혹시 로그인을 해서 접근되는 사이트일 경우)관리
#크롤링이 많으면 도중에 로그아웃(서버관리에서 비활동으로 감지)
#특정 게시물이 사라질경우 -> 팝업 발생(없는..) ->팝업처리 검토
#게시물 스캔시 -> 임계점 모름!!
#게시판을 스캔-> 메타정보 획득  ->loop를 돌려서 일괄적으로 방문 접근 처리
#searchModule.SetCategoryList(1, '')  스크립트 실행
#20은 임시값 게시물을 넘어갔을때 현상을 확인 하기 위해
for page in range(1, 21):
    try:
            #자바스크립트 구동하기
        driver.execute_script("searchModule.SetCategoryList(%s, '')"% page)
        #화면전환이 너무빨라 강제로 2초씩 지연
        time.sleep(2)
        print("%s 페이지 이동", page)
        ##########################################################
        # 여러 사이트에서 정보를 수집할 경우 공통 정보 정의 탄계 필요
        # 상품명, 코멘트, 기간1, 기간2, 가격, 평점, 썸네일(사진), 링크(상품상세정보)
        boxItems = driver.find_elements_by_css_selector('.oTravelBox>.boxList>li')
        #상품 하나 하나 접근
        for li in boxItems:
            print('상품명', li.find_element_by_css_selector('h5.proTit').text)
    except Exception as e1:
        print('오류', e1)





























```
